### purr

usando la función map obten los años bisiestos del siglo XX

```r
#función de repaso r
is_leap <- function(year){
    return((((year %% 4) == 0) && ((year %% 100)!=0)) || ((year %% 400)== 0))
}

SXX<-1901:2000
mask<-map(SXX,is_leap)
#str(mask)
leaps_SXX<-SXX[unlist(mask)]
leaps_SXX
```

### tidyr

```r
casos_diarios <-  read.csv(file = "Casos_Diarios_Estado_Nacional.csv")
#str(casos_diarios)
#length(names(casos_diarios))
caso_diario_list<-gather(casos_diarios,key="días",value="infectados",
                        4:856)
caso_diario_list
```

 ### DPLYR

```r
#Primera opción
#vect<- str_replace_all(caso_diario_list[4],"X","")
#str_replace_all(vect,"\\.","/")

# segunda opcion
#a<-caso_diario_list[4]%>% str_replace_all("X","")%>%str_replace_all("\\.","/")
#a

#Hacerlo función

normalizarfecha<-function(fecha){
    return(str_replace_all(fecha,"X","") %>% str_replace_all("\\.","/"))
    }
fecha=map(caso_diario_list[4],normalizarfecha)
#fecha

caso_diario_fecha<-cbind(caso_diario_list[c(T,T,T,F,T)],fecha)
caso_diario_fecha<-caso_diario_fecha[c(1,2,3,5,4)]
#caso_diario_fecha

salida<-separate(data=caso_diario_fecha,
                 col=días,
                 into=c("dia","mes","año"),
                 sep='/')
salida
```
